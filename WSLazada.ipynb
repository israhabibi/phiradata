```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup, SoupStrainer
import requests
import json
import time
```

```python
driver = webdriver.Chrome(executable_path='chromedriver')
data_prd = []
driver.get('https://www.lazada.co.id/beli-lemari-es/sharp/?spm=a2o4j.searchlistcategory.0.0.3a2b2dabXO7G5G&page=3')
html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')
kulkases = soup.find('div',id = 'root').find_all('div', class_ = 'c2prKC')
for kulkas in kulkases:
    namaProduk = kulkas.find('div', class_ = 'c16H9d').get_text()
    linkProduk = kulkas.find('div', class_ = 'c16H9d').find('a').get('href')
    hargaProduk = kulkas.find('span',class_ = 'c13VH6').get_text()
    try:
        hargaProdukCoret = kulkas.find('del',class_ = 'c13VH6').get_text()
    except:
        hargaProdukCoret = 'NaN'
    lokasi = kulkas.find('div',class_ = 'c15YQ9').get_text()
    try:
        penilaianProduk = kulkas.find('span',class_ = 'c3XbGJ').get_text()
    except:
        penilaianProduk = 'NaN'
    data_prd.append((namaProduk,linkProduk,hargaProduk,hargaProdukCoret, lokasi,penilaianProduk))
driver.quit()
```
